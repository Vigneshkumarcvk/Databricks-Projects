```python
df = spark.table("workspace.messy_ecommerce_db.silver_orders")
```

Creating Revenue Table


```python
from pyspark.sql.functions import col, sum, count
```

```python
gold_revenue = df.filter((col("status") == "Delivered") & (col("is_valid") == "True"))
```

Total Revenue


```python
revenue_kpi = gold_revenue.agg(sum("total_price").alias("total_revenue"))
display(revenue_kpi)
```

Revenue by Category

```python
category_revenue = gold_revenue.groupby("category").agg(sum("total_price").alias("revenue")).orderBy("revenue", ascending=False)
display(category_revenue)
```

```python
top_products = gold_revenue.groupBy("product").agg(sum("quantity").alias("total_quantity")).orderBy("total_quantity",ascending=False)
display(top_products)
```

```python
customer_spending = gold_revenue.groupBy("customer_name").agg(sum("total_price").alias("total_spend")).orderBy("total_spend",ascending=False)
display(customer_spending)
```

```python
order_status = df.groupBy("status").agg(count("*").alias("total_orders"))
display(order_status)
```

```python
category_revenue.write.mode("overwrite").format("delta").saveAsTable("workspace.messy_ecommerce_db.gold_category_revenue")
top_products.write.mode("overwrite").format("delta").saveAsTable("workspace.messy_ecommerce_db.gold_top_products")
customer_spending.write.mode("overwrite").format("delta").saveAsTable("workspace.messy_ecommerce_db.gold_customer_spending")
order_status.write.mode("overwrite").format("delta").saveAsTable("workspace.messy_ecommerce_db.gold_order_status")
```

```python
revenue_kpi.write.mode("overwrite").format("delta").saveAsTable("workspace.messy_ecommerce_db.gold_revenue_kpi")
```
